# Data Presentation
In data presentation i'll be cleaning the dataset for modelling and analysis

```python
# Let's check for dublicates
df.duplicated().sum()
```

```python
# Check if dublicates are still there
df=df.drop_duplicates()
df.duplicated().sum()
```

```python
# see the number of missing values per colunm
df.isnull().sum()
```

```python
df[['Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries','Total.Uninjured']]= df[['Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries','Total.Uninjured']].fillna(0)
df[['Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries','Total.Uninjured']]
```

```python
# replacing the missing values with 0 in the injury columns
injury_columns = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']
df[injury_columns] = df[injury_columns].fillna(0)
df[injury_columns].isnull().sum()
```

```python
# Replacing missing values with unknown in the categorical columns
categorical_columns = ['Broad.phase.of.flight','Event.Date','Location','Airport.Name','Injury.Severity']
df[categorical_columns] = df[categorical_columns].fillna('Unknown')
df[categorical_columns].isnull().sum()
```

```python
df[("Broad.phase.of.flight")].fillna("Unknown", inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Broad.phase.of.flight']
```

```python
import matplotlib.pyplot as plt
phase_counts=df['Broad.phase.of.flight'].value_counts()
plt.figure()
phase_counts.plot(kind='bar')
```

```python
# Dropping rows containg missing values in the following columns
df=df.dropna(subset=['Event.Date','Location','Airport.Name','Injury.Severity'])
```

Above i have cleaned specific columns from the dataset and left the rest with missing values.

# Accident Frequent by Aircraft Manufacturer

This visualization will help the business in looking at manufacturers with high cases of accidents in the past.

```python
# Creating a series with manufacturers as labels and accidents counts as values.
top_manufacturers = df['Make'].value_counts().head(10)


# Plotting the visualization
top_manufacturers.plot(kind='bar', figsize=(10, 6))
plt.title('Top 10 Aircraft Manufacturers with Highest Accidents')
plt.xlabel('Aircraft Manufacturer', fontsize=11)
plt.ylabel('Number of Accidents', fontsize=11)
plt.xticks(rotation=45)
plt.show()
```

# Accident trend overtime


A Download trend will suggest improving safety, making certain aircraft categories more attractive investments.

```python
# Convert Events.Date to a datetime object, extracts just the year and stores it in a new column Event. Year
df['Event.Year'] = pd.to_datetime(df['Event.Date'], errors='coerce').dt.year

# Group the accidents by the years and counts how many times they occured
accidents_by_year = df.groupby('Event.Year').size()

# plotting the line chart
accidents_by_year.plot(kind='line', figsize=(10, 6))
plt.title('Number of Accidents Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.show()
```

# Injury Severity Distribution

Aircraft associated with fewerfatal and serious injuries pose lower financial and reputational risk.

```python
# Selects injury columns and sums each column to get total injuries by severity level
severity_counts = df[['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']]
severity_totals = severity_counts.sum()

# Plotting a barchart to compare severity levels
severity_totals.plot(
    kind='bar',
    figsize=(8, 5),
    color=['red', 'orange', 'yellow', 'green']
)
plt.title('Distribution of injury severity')
plt.ylabel('Number of injuries')
plt.xlabel('')
# Remove xlabel as it's not meaningful for summed severity totals
plt.show()
```

## Summary of Analysis for Recommendations

Based on the performed data cleaning steps and visualizations, the following key insights can be drawn:

### 1. Top Aircraft Manufacturers with Highest Accidents
The 'Top 10 Aircraft Manufacturers with Highest Accidents' bar chart reveals that **Cessna** and **Piper** are by far the manufacturers associated with the highest number of reported accidents. It's notable that the data contains both capitalized and uncapitalized versions of these names (e.g., 'Cessna' and 'CESSNA', 'Piper' and 'PIPER'), suggesting a need for case-insensitive standardization for a more accurate consolidated count. Ignoring this, Cessna and Piper collectively account for a significant portion of accidents, which warrants further investigation into their specific aircraft models or operational contexts.

### 2. Accident Trend Over Time
The 'Number of Accidents Over Time' line chart indicates a general **decreasing trend** in the number of aviation accidents after a peak in the early 1980s. While there are fluctuations, the long-term pattern suggests improving safety standards or perhaps changes in reporting mechanisms. This downward trend is a positive indicator for overall aviation safety.

### 3. Distribution of Injury Severity
The 'Distribution of injury severity' bar chart highlights that the vast majority of individuals involved in aviation accidents are **uninjured** (441,889). Among those who sustain injuries, **fatalities** (50,201) are the most common severe outcome, followed by minor injuries (27,478) and then serious injuries (21,377). This distribution suggests that while accidents do occur, a significant proportion do not result in injury, and among those that do, fatal injuries are a primary concern.

### 4. Impact of Data Cleaning on Interpretations

*   **Missing Injury Values**: Filling missing values in `Total.Fatal.Injuries`, `Total.Serious.Injuries`, `Total.Minor.Injuries`, and `Total.Uninjured` columns with `0` was a pragmatic approach. This assumes that a lack of reported injury count implies no injuries occurred. If, however, some of these nulls represented unknown but existing injuries, the current injury severity distribution might slightly *underestimate* actual injury counts in some categories, particularly for less severe outcomes. For business decisions focused on minimizing fatalities, this assumption might not significantly skew the fatal injury count if underreporting is primarily for minor/uninjured cases.
*   **Missing Categorical Values**: Replacing missing values in `Broad.phase.of.flight`, `Event.Date`, `Location`, `Airport.Name`, and `Injury.Severity` with 'Unknown' allows for their inclusion in analyses. For example, a substantial 'Unknown' category in 'Broad.phase.of.flight' (27,782 entries) means that the insights derived from specific flight phases are based on a subset of the data, potentially obscuring patterns that occur in the 'Unknown' phase.
*   **Duplicate Removal**: The removal of 1390 duplicate rows ensures that accident counts and related statistics are not inflated by redundant entries, leading to more accurate analyses.

### Key Insights for Business Recommendations:

*   **Manufacturer Risk**: Cessna and Piper aircraft are involved in a disproportionately high number of accidents, suggesting these manufacturers (or specific models within their fleets) might require closer scrutiny regarding safety records, maintenance, or operational procedures.
*   **Safety Improvement**: The declining trend in overall accidents indicates that aviation safety has generally improved over time, which is a positive foundation for a new entrant in the industry.
*   **Injury Outcomes**: While accidents occur, a large percentage of individuals involved emerge uninjured. However, fatalities remain a significant concern among injured parties. Any new operations should prioritize aircraft and procedures that minimize fatal outcomes.
